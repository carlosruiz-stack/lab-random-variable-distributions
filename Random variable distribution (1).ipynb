```python
num = np.random.beta(a=4, b=15, size=300)

#create log-transformed data
data_log = np.log(num)

#define grid of plots
fig, axs = plt.subplots(nrows=1, ncols=2)

#create histograms
axs[0].hist(num, edgecolor='black')
axs[1].hist(data_log, edgecolor='black')

#add title to each histogram
axs[0].set_title('Original Data')
axs[1].set_title('Log-Transformed Data')
```

```python
# after we perform the transformation, we see that the plot takes the shape of a right-skewed distribution vs. the left-skewed distribution before the log transformation. 
```
